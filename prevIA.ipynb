# CRIACAO DA FUNCAO DE GERACAO DA MLP

```python
# Criar funcao para discretizar o valor float retornado pela rede
def discretizacao (x):
    if x >= 0.5:
        return 1
    else:
        return 0

def criaRNA(a, 
            b, 
            c, 
            d, 
            e, 
            f, 
            activationString, 
            numeroCamadas, 
            activationStringSaida, 
            o, 
            bt, 
            dados_treino_x_values, 
            dados_treino_y_values, 
            dados_teste_x_values, 
            dados_teste_y_values, 
            epochsValue,
            dataAfastado,
            dataNaoAfastado,
            dadosAfastadosTreino,
            dadosAfastadosTeste,
            dadosNaoAfastadosTreino,
            dadosNaoAfastadosTeste
           ):
    # Iniciando as variaveis
    model = None
    classes = None
    loss_and_metrics = None
    classesPredict = None
    crossTable = None
    crossTableOutPut = None
    
    # Informacao da operacao
    print('ID:'+ str(i) +' Otim:'+   str(o) +' Cam:'+ str(numeroCamadas) +' Bath:'+ str(bt) +' Neur:' + str(a) +'-'+ str(b) +'-'+ str(c) +'-'+ str(d) +'-'+ str(e) +'-'+ str(f))
    # Modelo Sequencial
    model = Sequential()
    # Adicao das camadas
    # Camada de entrada
    model.add(Dense(a, activation=activationString, input_dim=(shape-1)))
    # Camadas ocultas
    model.add(Dense(b, activation=activationString))
    if(c != 0):
        model.add(Dense(c, activation=activationString))
    if(d != 0):
        model.add(Dense(d, activation=activationString))
    if(e != 0):
        model.add(Dense(e, activation=activationString))
    if(f != 0):
        model.add(Dense(f, activation=activationString))

    # Camada de saída
    model.add(Dense(1, activation=activationStringSaida))
    # configurar o modelo 
    model.compile(loss=loss_function, optimizer=o, metrics=['accuracy'])
    # model.compile(loss='categorical_crossentropy', optimizer=o, metrics=['accuracy'])
    # treinar o modelo com os dados de treino
    history = model.fit(dados_treino_x_values, dados_treino_y_values, epochs=epochsValue, batch_size=bt, validation_data=(dados_teste_x_values, dados_teste_y_values))
    # print(history.history)
    # avaliar a acertividade do modelo
    loss_and_metrics = model.evaluate(dados_teste_x_values, dados_teste_y_values['foi_afastado_ano_previsao'], batch_size=bt)
      
    # fazer novas predicoes com os dados
    classes = model.predict(dados_teste_x_values, batch_size=bt)
    # Conversao dos dados em um pandas dataframe
    classesPredict = pd.DataFrame(classes, columns=['foi_afastado_previsao'])
    # criacao da coluna com os dados discretos
    classesPredict['discreto'] = classesPredict['foi_afastado_previsao'].apply(discretizacao)
    # matriz de confusao
    crossTable = confusion_matrix(dados_teste_y_values['foi_afastado_ano_previsao'], classesPredict['discreto'])
    # matriz para dataframe
    crossTableOutPut = pd.DataFrame(crossTable)
 
    timestamp = datetime.now()
    # gravacao do resultado no arquivo csv
    file = open('resultado/result_mlp_a_'+ str(anos) +'_anos_'+  str(o) + '_batch_' + str(bt) + '_camadas_' + str(numeroCamadas) +  '_date_' + str(date.today()) + '.csv',"a+")

    file.write( 
        str(i) +','+
        str(anos) +','+ 
        str(numeroCamadas) +','+ 
        str(o) +','+ 
        str(a) +','+ 
        str(b) +','+ 
        str(c) +','+ 
        str(d) +','+ 
        str(e) +','+
        str(f) +','+
        str(activationString) +','+ 
        str(1) +','+
        str(activationStringSaida) +','+
        str(loss_and_metrics[0]) +','+ 
        str(loss_and_metrics[1]) +','+
        str(bt) +','+
        str(epochsValue)  +','+
        str(crossTableOutPut.at[0,0]) +','+ 
        str(crossTableOutPut.at[0,1]) +','+ 
        str(crossTableOutPut.at[1,1]) +','+ 
        str(crossTableOutPut.at[1,0]) +','+ 
        str(crossTableOutPut.at[0,0] / (crossTableOutPut.at[0,0] + crossTableOutPut.at[0,1])) +','+ 
        str(crossTableOutPut.at[1,1] / (crossTableOutPut.at[1,1] + crossTableOutPut.at[1,0])) +','+
        # validation        
        str(history.history['val_loss'][9]) +','+
        str(history.history['val_accuracy'][9]) +','+
        str(history.history['loss'][9]) +','+
        str(history.history['accuracy'][9]) +','+
        # quantitativos
        str(dataAfastado.shape[0]) + ',' + 
        str(dataNaoAfastado.shape[0]) + ',' +
        str(dadosAfastadosTreino.shape[0]) +','+ 
        str(dadosAfastadosTeste.shape[0]) +','+ 
        str(dadosNaoAfastadosTreino.shape[0]) +','+ 
        str(dadosNaoAfastadosTeste.shape[0]) +','+ 
        str(dados_treino_x_values.shape[0]) + ',' +               
        str(dados_teste_x_values.shape[0]) +','+ 
        # horario
        str(timestamp)+','+ '\r\n')
    
    
    
    file.close()  
    # Tratamento do uso excessivo de memoria
    del model
    del classes
    del loss_and_metrics
    del classesPredict
    del crossTable
    del crossTableOutPut

    # Recolher o lixo pelo python
    garbage_collector.collect()
```
